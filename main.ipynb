Checker les NA

```python
# Liste des fichiers à vérifier
files_to_check = [
    'data/philly_business.json',
    'data/philly_reviews.json',
    'data/philly_users.json',
    'data/philly_tips.json',
    'data/philly_photos.json'
]

def check_missing_values(file_path):
    if not os.path.exists(file_path):
        print(f"Fichier introuvable : {file_path}")
        return

    print(f"\n--- Analyse de : {file_path} ---")
    
    # Chargement du fichier (format JSON Lines)
    try:
        df = pd.read_json(file_path, lines=True)
        
        # Calcul des NA
        missing_count = df.isnull().sum()
        total_rows = len(df)
        
        # Affichage uniquement des colonnes qui ont des manques
        if missing_count.sum() == 0:
            print("Aucune valeur manquante détectée")
        else:
            print(f"Total des lignes : {total_rows}")
            for col, count in missing_count.items():
                if count > 0:
                    percent = (count / total_rows) * 100
                    print(f"Colonne '{col}': {count} NA ({percent:.2f}%)")
    except Exception as e:
        print(f"Erreur lors de la lecture : {e}")

# Exécution du diagnostic
for file in files_to_check:
    check_missing_values(file)
```
